```python
import pandas as pd
import seaborn as srn
import statistics as sts
from numpy import nan
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
dataset = pd.read_csv('retail_store_sales.csv', sep=',')
dataset.head(15)
```

<h2>Tratamento dos dados</h2>
Tratamento: Category

```python
category_types = dataset.groupby(['Category']).size()
category_types
```

```python
dataset['Category'].isnull().sum()
```

Tratamento: True or False

```python
true_false = dataset.groupby(['Discount Applied']).size()
true_false
```

```python
dataset['Discount Applied'].isnull().sum()
```

```python
dataset['Discount Applied'].fillna('True', inplace=True)
dataset['Discount Applied'].isnull().sum()
```

```python
dataset.head(15)
```

Tratamento: Payment Method

```python
payment_method = dataset.groupby(['Payment Method']).size()
payment_method
```

```python
dataset['Payment Method'].isnull().sum()
```

Tratamento: Location

```python
location = dataset.groupby(['Location']).size()
location
```

```python
dataset['Location'].isnull().sum()
```

Tratamento: Item

```python
item = dataset.groupby(['Item']).size()
item
```

```python
dataset['Item'].isnull().sum()
```

```python
# identificar o item mencionado menos vezes
menor_valor = dataset['Item'].value_counts().tail(1)
```

```python
dataset['Item'].fillna(menor_valor.index[0], inplace=True)
dataset['Item'].isnull().sum()
```

```python
dataset.head(15)
```

Tratando: Total Spent

```python
print(type(dataset['Total Spent']))
```

```python
dataset['Total Spent'] = pd.to_numeric(dataset['Total Spent'], errors='coerce')
dataset['Total Spent'].isnull().sum()
```

```python
print(dataset['Total Spent'].head())
```

```python
max_quantity = dataset['Total Spent'].max()
max_quantity
```

```python
dataset['Total Spent'].fillna(max_quantity, inplace=True)
dataset['Total Spent'].isnull().sum()
```

```python
dataset.head(15)
```

Tratando: Transaction Date

```python
print(type(dataset['Transaction Date']))
```

```python
dataset.loc[(dataset['Transaction Date'].isin(['UNKNOWN', 'ERROR']))]
```

```python
dataset['Transaction Date'].isnull().sum()
```

Tratando: Price Per Unit

```python
dataset['Price Per Unit'] = pd.to_numeric(dataset['Price Per Unit'], errors='coerce')
dataset['Price Per Unit'].isnull().sum()
```

```python
max_quantity = dataset['Price Per Unit'].max()
max_quantity
```

```python
dataset['Price Per Unit'].fillna(max_quantity, inplace=True)
dataset['Price Per Unit'].isnull().sum()
```

```python
dataset.head(15)
```

Tratando: Transaction ID

```python
dataset[dataset.duplicated(['Transaction ID'], keep=False)]
```

Tratando: Quantity

```python
mediana = sts.median(dataset['Quantity'])
mediana
```

```python
dataset['Quantity'].fillna(mediana, inplace=True)
dataset.head(15)
```
