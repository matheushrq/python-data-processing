Substituindo ERROR e UNKNOWN para NaN

```python
dataset.loc[dataset['Total Spent'].isin(['ERROR', 'UNKNOWN']), 'Total Spent'] = None
media = dataset.groupby(['Total Spent']).size()
media
```

Considerando a coluna `Total Spent` que possui valores como 'ERROR' e 'UNKNOWN', voc precisar limpar esses valores antes de tentar convert-los para numrico. Uma forma de fazer isso  substitu-los por `NaN` e depois converter a coluna para um tipo numrico, e ento preencher os valores ausentes ou remov-los. Ou, como estvamos fazendo, substitu-los por um valor que j existe ou um que faa sentido, antes de converter.

```python
dataset['Total Spent'] = pd.to_numeric(dataset['Total Spent'], errors='coerce')
print(dataset['Total Spent'].head())
```

```python
mediana = sts.median(dataset['Total Spent'])
mediana
```

```python
dataset['Total Spent'].isnull().sum()
```

```python
dataset['Total Spent'].fillna(mediana, inplace=True)
dataset['Total Spent'].isnull().sum()
```
